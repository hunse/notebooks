Reproduce results from:

Rubin, D. B., Van Hooser, S. D., & Miller, K. D. (2015). The Stabilized
Supralinear Network: A Unifying Circuit Motif Underlying Multi-Input
Integration in Sensory Cortex. Neuron, 85(2), 402--417.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
from nengo.neurons import NeuronType
```

```python
# Define the neuron type

class MillerNeuron(NeuronType):
    def __init__(self, k, n, tau):
        self.k = k
        self.n = n
        self.tau = tau

    def rates(self, x, gain, bias):
        return self.k * np.maximum(x, 0)**self.n

    def gain_bias(self, max_rates, intercepts):
        gain = np.ones_like(max_rates)
        bias = np.zeros_like(intercepts)
        return gain, bias

    def step_math(self, dt, J, output, r):
        r_ss = self.k * np.maximum(J, 0)**self.n
        dr = -np.expm1(-dt / self.tau) * (r_ss - r)
        r += dr
        output[:] = r
        
        
from nengo.builder import Builder
from nengo.builder.signal import Signal
from nengo.builder.neurons import SimNeurons

@Builder.register(MillerNeuron)
def build_miller_neuron(model, neuron_type, neurons):
    model.sig[neurons]['rates'] = Signal(
        np.zeros(neurons.size_in), name="%s.rates" % neurons)
    model.add_op(SimNeurons(
        neurons=neuron_type,
        J=model.sig[neurons]['in'],
        output=model.sig[neurons]['out'],
        states=[model.sig[neurons]['rates']]))
```

```python
# test neuron model
neuron_type = MillerNeuron(k=0.04, n=2.0, tau=20e-3)
N = 50
J = np.linspace(-100, 100, N)

with nengo.Network() as model:
    u = nengo.Node(output=J)
    a = nengo.Ensemble(N, 1, neuron_type=neuron_type)
    nengo.Connection(u, a.neurons, synapse=None)
    ap = nengo.Probe(a.neurons)
    
sim = nengo.Simulator(model)
sim.run(0.1)

plt.subplot(121)
plt.plot(sim.trange(), sim.data[ap])
plt.subplot(122)
plt.plot(J, sim.data[ap][-1])
```

```python
# Figure 1: Normalization in a Nonlinear Ring Model
excitatory = MillerNeuron(k=0.04, n=2.0, tau=20e-3)
inhibitory = MillerNeuron(k=0.04, n=2.0, tau=10e-3)

def d(x, y):
    return np.abs(np.mod(x - y + 90, 180) - 90)

def G(x, y, sigma=1):
    return np.exp(-0.5 * d(x, y)**2 / sigma**2)

N = 180
dtheta = 180. / N
theta = dtheta * np.arange(N)

c = 50.
sigma_ori = 32.
sigma_FF = 30.
J_EE, J_IE, J_EI, J_II = 0.044, 0.042, 0.023, 0.018
Gtt = G(theta[:, None], theta[None, :], sigma=sigma_ori)
W_EE = J_EE * Gtt
W_IE = J_IE * Gtt
W_EI = -J_EI * Gtt
W_II = -J_II * Gtt

def make_ring():
    ring = nengo.Network(seed=9)
    with ring:
        #enc = nengo.Node(lambda x: G(x, theta, sigma=sigma_FF), size_in=1, size_out=N)
        E = nengo.Ensemble(N, 1, neuron_type=excitatory)
        I = nengo.Ensemble(N, 1, neuron_type=inhibitory)
        nengo.Connection(E.neurons, E.neurons, transform=W_EE)
        nengo.Connection(I.neurons, E.neurons, transform=W_EI)
        nengo.Connection(E.neurons, I.neurons, transform=W_IE)
        nengo.Connection(I.neurons, I.neurons, transform=W_II)
        Ep = nengo.Probe(E.neurons)
        Ip = nengo.Probe(I.neurons)
        
    ring.E = E
    ring.I = I
    ring.Ep = Ep
    ring.Ip = Ip
    return ring
    

plt.figure(figsize=(8,10))
    
# stimulus 1
with nengo.Network() as model:
    ring = make_ring()
    u1 = nengo.Node(c * G(45, theta, sigma=sigma_FF))    
    nengo.Connection(u1, ring.E.neurons)
    nengo.Connection(u1, ring.I.neurons)
    u1p = nengo.Probe(u1)
    
sim = nengo.Simulator(model)
sim.run(1.0)

plt.subplot(311)
plt.plot(theta, sim.data[u1p][-1], 'k--')
plt.plot(theta, sim.data[ring.Ep][-1], 'r')
plt.plot(theta, sim.data[ring.Ip][-1], 'b')
plt.ylim([0, 100])

# stimulus 2
with nengo.Network() as model:
    ring = make_ring()
    u1 = nengo.Node(c * G(135, theta, sigma=sigma_FF))    
    nengo.Connection(u1, ring.E.neurons)
    nengo.Connection(u1, ring.I.neurons)
    u1p = nengo.Probe(u1)
    
sim = nengo.Simulator(model)
sim.run(1.0)

plt.subplot(312)
plt.plot(theta, sim.data[u1p][-1], 'k--')
plt.plot(theta, sim.data[ring.Ep][-1], 'r')
plt.plot(theta, sim.data[ring.Ip][-1], 'b')
plt.ylim([0, 100])
    
# stimulus 1 and stimulus 2
with nengo.Network() as model:
    ring = make_ring()
    u1 = nengo.Node(c * G(45, theta, sigma=sigma_FF))
    u2 = nengo.Node(c * G(135, theta, sigma=sigma_FF))
    
    nengo.Connection(u1, ring.E.neurons)
    nengo.Connection(u1, ring.I.neurons)
    nengo.Connection(u2, ring.E.neurons)
    nengo.Connection(u2, ring.I.neurons)
    
    u1p = nengo.Probe(u1)
    u2p = nengo.Probe(u2)
    
sim = nengo.Simulator(model)
sim.run(1.0)

plt.subplot(313)
plt.plot(theta, sim.data[u1p][-1] + sim.data[u2p][-1], 'k--')
plt.plot(theta, sim.data[ring.Ep][-1], 'r')
plt.plot(theta, sim.data[ring.Ip][-1], 'b')
plt.ylim([0, 100])

# settling time for two unbalanced stimuli
c1, c2 = 45, 35

with nengo.Network() as model:
    ring = make_ring()
    u1 = nengo.Node(c1 * G(45, theta, sigma=sigma_FF))
    u2 = nengo.Node(c2 * G(135, theta, sigma=sigma_FF))
    
    nengo.Connection(u1, ring.E.neurons)
    nengo.Connection(u1, ring.I.neurons)
    nengo.Connection(u2, ring.E.neurons)
    nengo.Connection(u2, ring.I.neurons)
    
    u1p = nengo.Probe(u1)
    u2p = nengo.Probe(u2)
    
sim = nengo.Simulator(model)
sim.run(0.3)

fig = plt.figure(figsize=(20,10))
plt.subplot(221)
plt.plot(theta, sim.data[u1p][-1] + sim.data[u2p][-1], 'k--')
plt.plot(theta, sim.data[ring.Ep][-1], 'r')
plt.plot(theta, sim.data[ring.Ip][-1], 'b')
plt.ylim([0, 100])

plt.subplot(223)
plt.contourf(sim.trange(), theta, sim.data[ring.Ep].T)
plt.colorbar()

plt.subplot(224)
plt.contourf(sim.trange(), theta, sim.data[ring.Ip].T)
plt.colorbar()

#from mpl_toolkits.mplot3d import Axes3D
#ax = fig.add_subplot(212, projection='3d')
#ax.plot3D(sim.trange(), theta, sim.data[ring.Ep])
```

```python
# Figure 3: Spatial Contextual Interactions with Supralinear, Power-Law Input/Output Functions
def d(x, y):
    return np.abs(x - y)

def s(x, l):
    return (1. / (1 + np.exp(-(x + l/2.) / sigma_RF))) * (
        1 - 1. / (1 + np.exp(-(x - l/2.) / sigma_RF)))

c = 31

N = 101
dx = 1./3  # degrees
x = dx * np.arange(N)
x -= x.mean()

sigma_RF = 0.125 * dx
J_EE, J_IE = 1.0, 1.25
sigma_EE, sigma_IE = 2./3, 4./3  # degrees
W_EE = J_EE * G(x[:, None], x[None, :], sigma=sigma_EE)
W_IE = J_IE * G(x[:, None], x[None, :], sigma=sigma_IE)
W_EI = -1.0
W_II = -0.75

k, n = 0.01, 2.2
excitatory = MillerNeuron(k=k, n=n, tau=20e-3)
inhibitory = MillerNeuron(k=k, n=n, tau=10e-3)

def make_line():
    line = nengo.Network(seed=9)
    with line:
        E = nengo.Ensemble(N, 1, neuron_type=excitatory)
        I = nengo.Ensemble(N, 1, neuron_type=inhibitory)
        nengo.Connection(E.neurons, E.neurons, transform=W_EE)
        nengo.Connection(I.neurons, E.neurons, transform=W_EI)
        nengo.Connection(E.neurons, I.neurons, transform=W_IE)
        nengo.Connection(I.neurons, I.neurons, transform=W_II)
        Ep = nengo.Probe(E.neurons)
        Ip = nengo.Probe(I.neurons)
        
    line.E = E
    line.I = I
    line.Ep = Ep
    line.Ip = Ip
    return line


with nengo.Network() as model:
    line = make_line()
    u = nengo.Node(c * s(x, l=13))
    nengo.Connection(u, line.E.neurons)
    nengo.Connection(u, line.I.neurons)
    up = nengo.Probe(u)
    
sim = nengo.Simulator(model)
sim.run(1.0)

plt.subplot(111)
plt.plot(x, sim.data[up][-1], 'k--')
plt.plot(x, sim.data[line.Ep][-1], 'r')
plt.plot(x, sim.data[line.Ip][-1], 'b')
#plt.ylim([0, 100])
```

### Notes

In Figure 3 in the paper, all the firing rates have been normalized so that the maximum of the curve is 1.0, meaning that one is not sure of the absolute firing rates. In fact, for low contrast stimuli (e.g. $c = 6$), the firing rates are actually very small (around 1 Hz).
